```python
datos1 = [1,2,3,4,5]
datos2 = [5,4,3,2,1]

serie1 = pd.Series(datos1)
serie2 = pd.Series(datos2)

serie1
```

```python
serie2
```

con el metodo ***.lt*** podemos comparar si los elementos de la serie 1 es menor que los elementos de la serie 2

```python
serie1.lt(serie2)
```

el metodo ****.le *** sirve para comparar si los elementos de la serie 1 es menor o igual a los elementos de la serie dos

```python
serie1.le(serie2)
```

el metodo ****.eq *** sirve para comparar si los elementos de la serie 1 es igual a los elementos de la serie 2

```python
serie1.eq(serie2)
```

el metodo ****.gt *** sirve para comparar si los elementos de la serie 2 es mayor a los elementos de la serie 1

```python
serie1.gt(serie2)
```

el metodo ****.ge *** sirve para comparar si los elementos de la serie 2 es mayor o igual a los elementos de la serie 1

```python
serie1.ge(serie2)
```

```python
datos = list(range(1,6))

serie = pd.Series(datos)

serie
```

el metodo ***. where*** permite especificar una expresion booleana(condicion) donde aquellos elementos que pasen la condicion los convertira en objetos NAN ylos que cumplan la condicion mantendran su valor

```python
serie.where(serie > 1)
```

aqui le digo al metodo por cual valor va a sustituir los elementos que no cumplan en este caso con -1

```python
serie.where(serie > 1, -1)
```
