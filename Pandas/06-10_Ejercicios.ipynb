```python
datos = pd.Series(['100','200','python', '300.15','500.8'])

datos
```

```python
datos = pd.to_numeric(datos, errors='coerce')

datos
```

como vemos converti los datos de la serie a numerico pero los datos que no se pueden convertir a numericos con el argumento **errors=coerce*** los convierto a nan (datos no existente).

```python
datos = {'A':[1,2,3,4,5],
         'B':[9,8,7,6,5],
         'C':[2,3,5,7,1]}

datos
```

```python
type(datos)
```

***para ser un dataframe el diccionario debe tener en cada columa y fila la misma cantidad de datos***

```python
df= pd.DataFrame(datos)

df
```

```python
type(df)
```

*.iloc* sirve para que me de una columna en concreto

```python
columnaSerie = df.iloc[:,2]

columnaSerie
```

*.loc* sirve para que me de una fila en concreto

```python
filaSerie = df.loc[0,:]

filaSerie
```

```python
type(filaSerie)

type(columnaSerie)
```
